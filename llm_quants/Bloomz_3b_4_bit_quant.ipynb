```python
from transformers import AutoTokenizer, AutoModelForCausalLM,GPTQConfig
from transformers import __version__ as trans_version
import torch
```

```python
model = 'bigscience/bloomz-3b'
tokenizer = AutoTokenizer.from_pretrained(model)
gptq_config = GPTQConfig(bits=4, dataset = "c4", tokenizer=tokenizer)

model = AutoModelForCausalLM.from_pretrained(model,quantization_config=gptq_config, device_map='auto')
```

```python
prompt = """English: Hello there my name is Viraj.
French:"""
```

```python
from
```

```python
input = tokenizer(prompt, return_tensors="pt")
input.to("cuda")
output = model.generate(**input, max_length=len(prompt), num_beams=2, no_repeat_ngram_size=2, early_stopping=True)
```

```python
print(tokenizer.decode(output[0]))
```
